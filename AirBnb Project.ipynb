```python
import pandas as pd
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

# Download the stopwords dataset
nltk.download('stopwords')
nltk.download('punkt')

# Once downloaded, you can safely use the stopwords in your code
from nltk.corpus import stopwords

# Example of how to use stopwords
stops = set(stopwords.words('english'))
print(stops)
```

```python
def process_text(text):

    "Function to stem and remove stop words from a given text"
    # Tokenize text
    tokens = word_tokenize(text.lower())  # Convert to lower case
    # Stemming and stop words removal
    filtered_tokens = [stemmer.stem(word) for word in tokens if word not in stop_words and word.isalpha()]
    # Re-join tokens into a string
    return " ".join(filtered_tokens)

def clean_amenities(amenities):

    "Function to clean the amenities column, removing special characters.This version avoids using regular expressions."

    cleaned_amenities = amenities.replace('"', '').replace(',', ' ').replace('{', '').replace('}', '')
    # Split based on spaces and rejoin to ensure single spacing
    cleaned_amenities = ' '.join(cleaned_amenities.split())
    return cleaned_amenities

# Initialize the Porter Stemmer and get English stop words
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

# Clean and process the amenities column
airbnb_df['amenities'] = airbnb_df['amenities'].apply(clean_amenities).apply(process_text)


airbnb_df['description'] = airbnb_df['description'].apply(process_text)


# Merge the processed amenities and description columns
airbnb_df['merged_amenities_description'] = airbnb_df['amenities'] + " " + airbnb_df['description']
```

## BUILDING THE PREDICTIVE MODELS


## Step 1: Develop the Individual Models
### Linear Regression Model (Price Prediction):

Inputs: Numerical variables related to the Airbnb listing (such as location, number of rooms, amenities, etc.).
Output: Predicted price of the listing.

### Logistic Regression Model (Occupancy Prediction):

Inputs: Date information (month, week, day) and the average daily rate (ADR), which will include the predicted price from the linear model.
Output: Probability of the listing being rented on the given day.

### Step 2: Combine the Models in an Ensemble
Input: Date information and initial listing data.
Process:
Use the listing data to predict the price with the linear regression model.
Take the predicted price and combine it with the date information as inputs for the logistic regression model.
Predict the likelihood of occupancy using the logistic regression model.

## Model Training and Evaluation:

1. Train Regression using Linear Regression, Decision Tree Regressors, and Random Forest Regressors for baseline performance.
2. Train Classification using Logistic Regression, Decision Tree Classifiers, and Random Forest classifiers for baseline performance.


```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, roc_auc_score, accuracy_score, roc_curve
from sklearn.preprocessing import StandardScaler

# Sample a portion of the Airbnb dataset
sampled_airbnb_df = airbnb_df.sample(frac=0.35, random_state=42)  # Sample 50% of the data

# Sample a portion of the ADR dataset
sampled_adr_df = adr_df.sample(frac=0.35, random_state=42)  # Sample 50% of the data

# Prepare data for linear regression on the sampled Airbnb dataset
sampled_airbnb_df = sampled_airbnb_df[sampled_airbnb_df['log_price'] > 0]  # Filter out non-positive price values
X_regression = airbnb_df.drop(columns=['log_price'])
y_regression_log = airbnb_df['log_price']

# Split the data
X_train_reg, X_test_reg, y_train_reg_log, y_test_reg_log = train_test_split(X_regression, y_regression_log, test_size=0.3, random_state=42)

# Train linear regression model
linear_reg = LinearRegression()
linear_reg.fit(X_train_reg, y_train_reg_log)

# Predict on test set
y_pred_reg_log = linear_reg.predict(X_test_reg)

# Convert predictions back to original scale
y_pred_reg = np.exp(y_pred_reg_log)
y_test_reg = np.exp(y_test_reg_log)

# Calculate error metrics for linear regression
mae_lr = mean_absolute_error(y_test_reg, y_pred_reg)
mse_lr = mean_squared_error(y_test_reg, y_pred_reg)
rmse_lr = sqrt(mse_lr)

# Print error metrics for linear regression
print("Linear Regression:")
print(f"MAE: {mae_lr}, MSE: {mse_lr}, RMSE: {rmse_lr}")

# Plot predicted vs actual prices for linear regression
plt.figure(figsize=(8, 6))
plt.scatter(y_test_reg, y_pred_reg, color='blue', alpha=0.5)
plt.plot([min(y_test_reg), max(y_test_reg)], [min(y_test_reg), max(y_test_reg)], color='red', linestyle='--')
plt.title('Linear Regression: Actual vs Predicted Prices')
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.show()

# Prepare data for logistic regression on the sampled ADR dataset
sampled_adr_df['adr_standardized'] = StandardScaler().fit_transform(sampled_adr_df[['adr']])
sampled_adr_df['is_occupied'] = 1 - sampled_adr_df['is_canceled']

X_classification = sampled_adr_df[['adr_standardized', 'arrival_date_year', 'arrival_date_month', 'arrival_date_week_number', 'arrival_date_day_of_month']]
X_classification['arrival_date_month'] = pd.to_datetime(X_classification['arrival_date_month'], format='%B').dt.month  # Convert month to numeric
y_classification = sampled_adr_df['is_occupied']

# Z-score normalization for features
scaler_classification = StandardScaler()
X_classification_standardized = scaler_classification.fit_transform(X_classification)

# Split the data
X_train_class, X_test_class, y_train_class, y_test_class = train_test_split(X_classification_standardized, y_classification, test_size=0.3, random_state=42)

# Prepare classification models
classification_models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree Classifier': DecisionTreeClassifier(),
    'Random Forest Classifier': RandomForestClassifier()
}

# Evaluate classification models
for name, model in classification_models.items():
    model.fit(X_train_class, y_train_class)
    y_pred_class = model.predict(X_test_class)
    y_prob_class = model.predict_proba(X_test_class)[:, 1] if hasattr(model, "predict_proba") else None
    precision = precision_score(y_test_class, y_pred_class)
    recall = recall_score(y_test_class, y_pred_class)
    roc_auc = roc_auc_score(y_test_class, y_prob_class) if y_prob_class is not None else None
    accuracy = accuracy_score(y_test_class, y_pred_class)
    print(f"{name} - Precision: {precision}, Recall: {recall}, ROC AUC: {roc_auc}, Accuracy: {accuracy}")

    # Visualize ROC curve
    if y_prob_class is not None:
        fpr, tpr, thresholds = roc_curve(y_test_class, y_prob_class)
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
        plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve - {name}')
        plt.legend(loc="lower right")
        plt.show()

# Prepare regression models
regression_models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree Regressor': DecisionTreeRegressor(),
    'Random Forest Regressor': RandomForestRegressor()
}

# Evaluate regression models
for name, model in regression_models.items():
    model.fit(X_train_reg, y_train_reg_log)
    y_pred_reg_log = model.predict(X_test_reg)
    y_pred_reg = np.exp(y_pred_reg_log)
    y_test_reg = np.exp(y_test_reg_log)

    mae = mean_absolute_error(y_test_reg, y_pred_reg)
    mse = mean_squared_error(y_test_reg, y_pred_reg)
    rmse = sqrt(mse)

    print(f"{name}: MAE: {mae}, MSE: {mse}, RMSE: {rmse}")

    plt.figure(figsize=(8, 6))
    plt.scatter(y_test_reg, y_pred_reg, color='blue', alpha=0.5)
    plt.plot([min(y_test_reg), max(y_test_reg)], [min(y_test_reg), max(y_test_reg)], color='red', linestyle='--')
    plt.title(f'{name}: Actual vs Predicted Prices')
    plt.xlabel('Actual Prices')
    plt.ylabel('Predicted Prices')
    plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Preprocess and split Airbnb data
X_airbnb = sampled_airbnb_df.drop(columns=['log_price'])
y_airbnb = np.exp(sampled_airbnb_df['log_price'])  # Exponentiate to transform log price back to normal price
X_train_airbnb, X_test_airbnb, y_train_airbnb, y_test_airbnb = train_test_split(X_airbnb, y_airbnb, test_size=0.2, random_state=42)

# Fit the RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100, max_depth=20, random_state=42)
regressor.fit(X_train_airbnb, y_train_airbnb)

# Preprocess Hotel data
adr_df['arrival_date_month'] = adr_df['arrival_date_month'].astype(int)  # Corrected to handle integer month values
adr_df['is_occupied'] = 1 - adr_df['is_canceled']
X_adr = adr_df[['adr', 'arrival_date_month', 'arrival_date_day_of_month']].copy()

# Scale the 'adr' feature
scaler_adr = StandardScaler()
X_adr['adr'] = scaler_adr.fit_transform(X_adr[['adr']])
# Define y_adr
y_adr = adr_df['is_occupied']

# Split the data for classification
X_train_adr, X_test_adr, y_train_adr, y_test_adr = train_test_split(X_adr, y_adr, test_size=0.2, random_state=42)

# Fit the RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100, max_depth=20, random_state=42)
classifier.fit(X_train_adr, y_train_adr)

# Ask user for input month and day
month_input = int(input("Enter the month (1-12): "))
day_input = int(input("Enter the day of the month (1-31): "))

# Get prices predicted by the regressor for test set
predicted_prices = regressor.predict(X_test_airbnb)

# Scale these predicted prices using the ADR scaler
scaled_prices = scaler_adr.transform(predicted_prices.reshape(-1, 1))

# Calculate residual errors for each prediction
residual_errors = y_test_airbnb - predicted_prices

# Predict occupancy using the classification model
input_features = np.hstack([scaled_prices, np.full((len(scaled_prices), 1), month_input), np.full((len(scaled_prices), 1), day_input)])
predicted_occupancy = classifier.predict_proba(input_features)[:, 1]

# Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(predicted_prices, predicted_occupancy, c=residual_errors, cmap='coolwarm', alpha=0.6)
plt.title('Predicted Prices vs. Occupancy Probability')
plt.xlabel('Predicted Prices ($)')
plt.ylabel('Probability of Occupancy (%)')
plt.colorbar(label='Residual Error')
plt.grid(True)
plt.show()

# Print a summary including residual errors
print("Summary of Predicted Prices, Occupancy, and Residual Errors:")
for price, occupancy, error in zip(predicted_prices, predicted_occupancy, residual_errors):
    print(f"Predicted Price: ${price:.2f}, Predicted Occupancy Probability: {occupancy:.2%}, Residual Error: {error:.2f}")
```
